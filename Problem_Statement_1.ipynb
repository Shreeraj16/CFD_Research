IMPLICIT METHOD 1D DIFFUSION EQUATION

```python
#IMPLICIT METHOD 1D DIFFUSION EQUATION
import math as mt
import numpy as np
import matplotlib.pyplot as plt 
import time

#lets take the input of the numzber of points you want your mesh to get divided into
mesh_points = int(input("Enter the number of the mesh points: "))

#here our given domain is [0,1] hence lets start making the grids 
x_cor = np.array(np.linspace(0,1,mesh_points))

#lets start initializing the required variables for the given problem

alpha = 1 
dx = 1 / (mesh_points-1)
lamda = 0.4
dt = (lamda*(dx**2))/alpha
print(dt)
t_end=int(input("Enter the end time: "))
t=dt

u_old = np.zeros(mesh_points) #creating an zero array for storing the data as the nth time step
u_new = np.zeros(mesh_points) #creating an zero array for storing the data as the n+1th time step
steady_term = np.zeros(mesh_points) #here we are creating the array for storing the analytical solution of the differential equation which is given by the equation f(x-ct) where f(x) is the initial condition

#we will start storing the inital conditions in the array 
for i in range(0,mesh_points):
  if (abs(x_cor[i]-0.5)>0.1):
    u_old[i]=0
  else:
    u_old[i]=1

s = alpha * dt/(dx**2.0) 
# Diagonal elements of system matrix
d  = np.zeros(mesh_points)        # main diagonal elements
u  = np.zeros(mesh_points)        # upper diagonal
l  = np.zeros(mesh_points)        # lower diagonal

# Construct the elements of the coefficient matrix
l[0]  = 0.0
u[0]  = 0.0
d[0]  = 1.0
l[-1] = 0.0
u[-1] = 0.0
d[-1] = 1.0
for i in range(1,mesh_points-1):
  d[i] =  1.0+2.0*s
  l[i] = -s
  u[i] = -s

def tdma(num, dia, upp, low, uPrev, u):

  P    = np.zeros(num)
  Q    = np.zeros(num)

  # Forward elimination
  P[0] = -upp[0]/dia[0]
  Q[0] =  uPrev[0]/dia[0]
  for i in range(1, mesh_points):
    P[i] = -(upp[i]/(dia[i]+low[i]*P[i-1]))
    Q[i] =  (uPrev[i] - low[i]*Q[i-1])/(dia[i]+low[i]*P[i-1])
    
  # Backward Substitution 
  u[-1] = Q[-1]
  for i in range(mesh_points-2,-1,-1):
    u[i] = P[i]*u[i+1]+Q[i]
  return u

# Solve the discretized equation until steady state is achieved
norm        = 100.0       # Measure of residual or error
iter        = 0           # Iteration counter
converged   = False       # Convergence flag - changes to "True" upon convergence

while converged==False:
  #Lets start the calculation of the new time step
  u_new = tdma(mesh_points, d, u, l, u_old, u_new) 
  u_new[0]=0
  u_new[-1]=0
  t=t+dt #incrementing the time of the solution  
  for i in range(1, mesh_points-1):
    steady_term[i] = abs(alpha*(u_new[i+1] - 2.0*u_new[i] + u_new[i-1]) / (dx**2.0))
  norm = max(steady_term)
  if norm <0.00001:
    converged = True

  # Increment iteration counter
  iter = iter + 1

  # Print norm to monitor convergence
  #print(iter, norm)
    
  for i in range(0, mesh_points):
    u_old[i] = u_new[i]
  print(u_new)

plt.plot(x_cor,u_new)
plt.xlabel("Length of Rod")
plt.ylabel("Velocity (u)")
plt.show()


```

```python
#IMPLICIT METHOD 1D DIFFUSION EQUATION using gauss siedel
import math as mt
import numpy as np
import matplotlib.pyplot as plt 
import time

def gauss_seidel(a, x ,b): #we are using gauss - siedel method for the approximations of the answer
  #Finding length        
  n = len(a)                   
  # for loop for 3 times as to calculate x, y , z
  for j in range(0, n):        
    # temp variable d to store b[j]
    d = b[j]                  
          
    # to calculate respective xi, yi, zi
    for i in range(0, n):     
      if(j != i):
        d-=a[j][i] * x[i]
        # updating the value of our solution        
    x[j] = d / a[j][j]
    # returning our updated solution           
  return x  

#lets take the input of the number of points you want your mesh to get divided into
mesh_points = int(input("Enter the number of the mesh points: "))

#here our given domain is [0,1] hence lets start making the grids 
x_cor = np.array(np.linspace(0,1,mesh_points))

#lets start initializing the required variables for the given problem

alpha = 1 
dx = 1 / (mesh_points-1)
lamda = 0.4
dt = (lamda*(dx**2))/alpha
t=dt
ts= 1
tsteps=[ts]

u_old = np.zeros(mesh_points) #creating an zero array for storing the data as the nth time step
u_new = np.zeros(mesh_points) #creating an zero array for storing the data as the n+1th time step
steady_term = np.zeros(mesh_points) #here we are creating the array for storing the analytical solution of the differential equation which is given by the equation f(x-ct) where f(x) is the initial condition

#we will start storing the inital conditions in the array 
for i in range(0,mesh_points):
  if (abs(x_cor[i]-0.5)>0.1):
    u_old[i]=0
  else:
    u_old[i]=1
#forming the coefficient matrix

s = alpha * dt/(dx**2.0) 

A = np.zeros((mesh_points, mesh_points))
# Construct the elements of the coefficient matrix
for i in range (1,mesh_points-1):      # For the inner nodes
    A[i][i]   = 1.0+2.0*s
    A[i][i-1] = -s
    A[i][i+1] = -s
A[0][0]     = 1.0               # For boundary node  
A[-1][-1]   = 1.0               # For boundary node

# Solve the discretized equation until steady state is achieved
norm        = 100.0       # Measure of residual or error
iter        = 0           # Iteration counter
converged   = False       # Convergence flag - changes to "True" upon convergence
norm_collect=[]
while converged==False:
  #Lets start the calculation of the new time step
  u_new = np.zeros(mesh_points)
  p = np.zeros(mesh_points)
  for i in range (0,mesh_points):
    p[i]=u_old[i]
  for i in range(0, 25):            
    u_new = gauss_seidel(A, u_new, p)
  u_new[0]=0
  u_new[-1]=0
  t=t+dt #incrementing the time of the solution  
  for i in range(1, mesh_points-1):
    steady_term[i] = abs(alpha*(u_new[i+1] - 2.0*u_new[i] + u_new[i-1]) / (dx**2.0))
  norm = max(steady_term)
  norm_collect.append(norm)
  if norm <0.00001:
    converged = True

  # Increment iteration counter
  iter = iter + 1

  # Print norm to monitor convergence
  print(iter, norm)
  ts = ts + 1
  tsteps.append(ts)  
  for i in range(0, mesh_points):
    u_old[i] = u_new[i]

plt.plot(x_cor,u_new)
plt.xlabel("Length of Rod")
plt.ylabel("Velocity (u)")
plt.show()

tsteps = np.linspace(1,iter,iter)
norm_collect = np.array(norm_collect)
plt.plot(tsteps,norm_collect)
plt.yscale('log')
plt.xlabel("time steps")
plt.ylabel("norm")
plt.show()
```

```python
#IMPLICIT METHOD 1D DIFFUSION EQUATION
import math as mt
import numpy as np
import matplotlib.pyplot as plt 
import time

#lets take the input of the numzber of points you want your mesh to get divided into
mesh_points = int(input("Enter the number of the mesh points: "))

#here our given domain is [0,1] hence lets start making the grids 
x_cor = np.array(np.linspace(0,1,mesh_points))

#lets start initializing the required variables for the given problem

alpha = 1 
dx = 1 / (mesh_points-1)
lamda = 0.4
dt = (lamda*(dx**2))/alpha
t=dt

u_old = np.zeros(mesh_points) #creating an zero array for storing the data as the nth time step
u_new = np.zeros(mesh_points) #creating an zero array for storing the data as the n+1th time step
steady_term = np.zeros(mesh_points) #here we are creating the array for storing the analytical solution of the differential equation which is given by the equation f(x-ct) where f(x) is the initial condition

#we will start storing the inital conditions in the array 
for i in range(0,mesh_points):
  if (abs(x_cor[i]-0.5)>0.1):
    u_old[i]=0
  else:
    u_old[i]=1

s = alpha * dt/(dx**2.0) 
# Diagonal elements of system matrix
d  = np.zeros(mesh_points)        # main diagonal elements
u  = np.zeros(mesh_points)        # upper diagonal
l  = np.zeros(mesh_points)        # lower diagonal

# Construct the elements of the coefficient matrix
l[0]  = 0.0
u[0]  = 0.0
d[0]  = 1.0
l[-1] = 0.0
u[-1] = 0.0
d[-1] = 1.0
for i in range(1,mesh_points-1):
  d[i] =  1.0+2.0*s
  l[i] = -s
  u[i] = -s

def tdma(num, dia, upp, low, uPrev, u):

  P    = np.zeros(num)
  Q    = np.zeros(num)

  # Forward elimination
  P[0] = -upp[0]/dia[0]
  Q[0] =  uPrev[0]/dia[0]
  for i in range(1, mesh_points):
    P[i] = -(upp[i]/(dia[i]+low[i]*P[i-1]))
    Q[i] =  (uPrev[i] - low[i]*Q[i-1])/(dia[i]+low[i]*P[i-1])
    
  # Backward Substitution 
  u[-1] = Q[-1]
  for i in range(mesh_points-2,-1,-1):
    u[i] = P[i]*u[i+1]+Q[i]
  return u

# Solve the discretized equation until steady state is achieved
norm_collect = []         # To collect the norm 
sub_list = list([])       # Making a list to store the difference between the two arrays
norm        = 100.0       # Measure of residual or error
iter        = 0           # Iteration counter
converged   = False       # Convergence flag - changes to "True" upon convergence


while (norm >10**-6):
  #Lets start the calculation of the new time step
  u_new = tdma(mesh_points, d, u, l, u_old, u_new) 
  u_new[0]=0
  u_new[-1]=0
  t=t+dt #incrementing the time of the solution  
  sub_list = list([])
  for i1,i2 in zip(u_new,u_old):
    sub_list.append(abs(i1-i2))
  norm = max(sub_list)/dt
  norm_collect.append(norm)


  # Increment iteration counter
  iter = iter + 1

  # Print norm to monitor convergence
  #print(iter, norm)
    
  for i in range(0, mesh_points):
    u_old[i] = u_new[i]
  #print(u_new)

plt.plot(x_cor , u_new)
plt.xlabel("Length of Rod")
plt.ylabel("Velocity (u)")
plt.show()
norm_collect = np.array(norm_collect)
iterations = np.linspace(1,iter,iter)
plt.plot(iterations , norm_collect)
plt.xlabel("Iterations")
plt.yscale('log')
plt.ylabel("Error in logarithmic scale")
plt.show()
print(u_new)
```
